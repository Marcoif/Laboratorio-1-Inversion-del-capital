## Bajar los datos de Yahoo finance para ver las cotizaciones que tienen desde la fecha que se requiere hasta el final del periodo solicitado ##

```python
prices= fn.prices_download(global_tickers)
prices.head()
```

### Al estar revisando los datos de descargas y los datos que se tenian dentro de los csv me pude percatar que no estan cuadrados correctamente, ya que los archivos no concuerdan con los precios de cierrre de las acciones una vez descargadas de yahoofinance, es por ello que se creo esta funcion interna para que pueda cuadrar cronologicamente estos datos ###

```python
dates_list = [datetime.datetime.strptime(i, "%Y%m%d").date() for i in fechas['index_fechas']]
for i in range(len(fechas['index_fechas'])):
    for j in range(31):
        if fechas['index_fechas'][i] in dt.filenames[j]:
            dt.data_archivos[fechas['index_fechas'][i]]=dt.data_archivos.pop(dt.filenames[j])
            dt.data_archivos[dates_list[i]]=dt.data_archivos.pop(fechas['index_fechas'][i])
```

<font size = 3> Aqui ya se observa los tickers de yahoo finance como deberian de estar y en orden para que ahora si los pesos coincidan con los solicitados en el primer archivo del csv


```python
dt.data_archivos[dates_list[0]]
```

<br>

<font size = 4> Al tener los datos de yahoofinance de forma diaria solo nos interesan los precios de cierre en las fechas especificas de cada archivo, es por ello que se creo la funcion clean_prices la cual solo nos muestra esos precios en las fechas solicitadas de todos los tickers

```python
clean_prices, data_close = fn.clean_price(data_archivos=dt.data_archivos, dates_list=dates_list, data=prices, t_fechas=fechas['t_fechas'])
clean_prices.head()
```

```python
pos_datos = fn.pasiva_ini(data_archivos=dt.data_archivos, dates_list=dates_list, precios=clean_prices)
pos_datos.head(10)
```
